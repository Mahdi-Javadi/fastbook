```python
from pathlib import Path

path = Path()
```

```python
path.cwd()
cur_dir = Path.cwd()
print(f"cur_dir: {cur_dir}")
print(f"two-level up: {Path('../..').resolve()}")
print(f"two-level up with parent: {cur_dir.parent.parent}")
print(f"expanduser method: {Path("~/git-repos").expanduser()}")
print(f"home method: {Path.home() / "git-repos"}")

## Print files in a directory
print(f"{10 * "=="}\nFiles in the directory:\n")
for p in (Path.home() / "git-repos").iterdir():
    print(p.name)

# The glob method finds all the pathnames matching a specified pattern.
# "glob" stands for "global", referring to global pattern matching (using wildcards like * and ?).
print(f"\njson files in git-repos directory:")
for json_file in (Path.home() / "git-repos").glob("*.json"):
    print(json_file.name)
    with json_file.open() as f:
        print(f.read())

# We can create files and folders using the Path class using the touch and mkdir methods.
tmp_folder ="/tmp/dummy_folder" 
dummy_folder = Path(tmp_folder)
dummy_folder.mkdir(exist_ok=True)
dummy_file = (dummy_folder / "dummy_file.txt").touch()
for p in Path(tmp_folder).iterdir():
    print(p.name)
```
